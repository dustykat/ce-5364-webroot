# **Course Outline: Groundwater Contaminant Fate and Transport Modeling**

## **Course Description**
This graduate-level course explores the fate and transport of contaminants in groundwater through analytical and numerical modeling. The course covers theoretical foundations, regulatory frameworks, and remediation strategies. Students will apply analytical models using Jupyter Notebooks and Spreadsheets and develop numerical models using MODFLOW6 with FLOPY as the primary interface.

## **Course Structure**
- **Meeting Schedule:** 24 sessions (Each session ~1.5–2 hours)
- **Primary Learning Tools:** Jupyter Notebooks, Spreadsheets, MODFLOW6, FLOPY
- **Textbooks:** Open-access groundwater modeling and contaminant transport resources
- **Assessment Methods:** Homework, modeling projects, case studies, and exams

---

## **Course Outline**

### **Module 1: Introduction to Groundwater Contaminant Transport (Sessions 1–4)**
1. **Introduction to Groundwater Flow and Contaminant Transport**
   - Importance of contaminant transport modeling
   - Overview of groundwater flow equations
   - Advection, dispersion, diffusion, sorption, and degradation
   
2. **Conceptual Models of Contaminant Fate and Transport**
   - Sources of contamination
   - Hydrogeologic framework and transport pathways
   - Role of retardation and degradation
   
3. **Regulatory Overview and Risk Assessment** (Reserved session for regulatory issues)
   - U.S. and international groundwater contamination regulations
   - Risk-based corrective actions and exposure pathways
   - Case studies of regulatory decisions
   
4. **Introduction to Modeling Environments**
   - Overview of Jupyter Notebooks and Spreadsheets for analytical modeling
   - Introduction to MODFLOW6 and FLOPY
   - Comparison of analytical and numerical approaches

---

### **Module 2: Analytical Modeling of Contaminant Transport (Sessions 5–10)**
5. **One-Dimensional Transport with Advection and Dispersion**
   - Derivation of the Advection-Dispersion Equation (ADE)
   - Analytical solutions using Jupyter Notebooks and Spreadsheets
   
6. **One-Dimensional Transport with Sorption and Decay**
   - Retardation factor and first-order decay
   - Example problems with parameter sensitivity analysis
   
7. **Two-Dimensional Contaminant Transport**
   - Instantaneous vs. continuous sources
   - Application of the Ogata-Banks solution
   
8. **Plume Evolution and Well Capture Zones**
   - Estimating plume lengths and wellhead protection areas
   - Analytical tools for well capture assessment
   
9. **Heterogeneity and Stochastic Approaches in Analytical Models**
   - Effects of heterogeneity on contaminant transport
   - Monte Carlo simulations for uncertainty analysis
   
10. **Case Study: Using Analytical Models for Risk Assessment**
    - Real-world contaminant transport scenario
    - Group work on applying analytical solutions

---

### **Module 3: Numerical Modeling with MODFLOW6 and FLOPY (Sessions 11–18)**
11. **Introduction to MODFLOW6 and FLOPY**
    - Overview of numerical groundwater flow modeling
    - Installation and setup of MODFLOW6 and FLOPY
    
12. **Building a Simple MODFLOW Model in FLOPY**
    - Setting up a steady-state groundwater flow model
    - Visualization of results in Python
    
13. **Solute Transport Modeling in MODFLOW (MT3DMS and MODFLOW6)**
    - Coupling MODFLOW with transport models
    - Setting up transport simulations in FLOPY
    
14. **Boundary Conditions and Source Terms**
    - Representation of wells, recharge, and contamination sources
    - Sensitivity of boundary conditions on transport outcomes
    
15. **Calibrating Transport Models**
    - Parameter estimation and model calibration techniques
    - Model validation using observed data
    
16. **Advanced Topics: Reactive Transport and Multi-Species Modeling**
    - Biodegradation, redox reactions, and multi-species fate
    - Case studies with reactive transport models
    
17. **Numerical Model Uncertainty and Sensitivity Analysis**
    - Parameter uncertainty and sensitivity analysis techniques
    - Use of PEST for automated calibration
    
18. **Case Study: Applying MODFLOW6/FLOPY to a Contaminated Site**
    - Hands-on group project with real-world data
    - Model setup, calibration, and transport simulation

---

### **Module 4: Groundwater Remediation Technologies and Applications (Sessions 19–24)**
19. **Introduction to Groundwater Remediation**
    - Overview of remediation technologies (pump-and-treat, in situ bioremediation, permeable reactive barriers)
    - Selecting appropriate remediation strategies
    
20. **Remediation System Design and Performance**
    - Modeling remediation strategies using transport models
    - Predicting cleanup times and monitoring well placement
    
21. **Regulatory Session: Remediation Case Studies and Legal Frameworks**
    - Environmental cleanup regulations and risk-based decision-making
    - Case studies of successful and failed remediation efforts
    
22. **Integration of Analytical and Numerical Models for Decision Making**
    - Hybrid modeling approaches in groundwater management
    - When to use analytical vs. numerical models
    
23. **Final Project Workshop**
    - Student presentations of modeling projects
    - Peer feedback and discussion on lessons learned
    
24. **Final Exam / Take-Home Assessment**
    - Exam covering theoretical concepts and applied modeling

---

## **Assessment & Grading**
- **Homework Assignments (30%)** – Analytical and numerical modeling exercises
- **Midterm Exam (20%)** – Covers theoretical foundations and modeling principles
- **Final Project (30%)** – Applied modeling of a groundwater contamination case study
- **Class Participation (10%)** – Discussions, peer reviews, and workshops
- **Final Exam (10%)** – Cumulative assessment of concepts and applications

---

## **Software & Tools**
- **Jupyter Notebooks** – Python-based analytical modeling
- **Microsoft Excel / Google Sheets** – Spreadsheet-based solutions
- **MODFLOW6 / FLOPY** – Numerical modeling of groundwater flow and transport
- **PEST** – Model calibration and parameter estimation

This course outline balances theoretical foundations, hands-on modeling, regulatory considerations, and real-world applications to provide students with a comprehensive understanding of groundwater contaminant fate and transport modeling.
